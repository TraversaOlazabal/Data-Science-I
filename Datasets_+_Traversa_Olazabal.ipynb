```python
import pandas as pd
```

```python
# Enlace directo al archivo en GitHub (archivo de ajedrez)
chess_url = 'https://raw.githubusercontent.com/TraversaOlazabal/DataScience-Datasets/refs/heads/main/games%20chess.csv'
chess_data = pd.read_csv(chess_url)
print("Chess Dataset:")
print(chess_data.head())

# Enlace directo al archivo de Bitcoin en GitHub
bitcoin_url = 'https://raw.githubusercontent.com/TraversaOlazabal/DataScience-Datasets/refs/heads/main/btc_2015_2024.csv'
bitcoin_data = pd.read_csv(bitcoin_url)
print("\nBitcoin Dataset:")
print(bitcoin_data.head())
```

```python
print("Tamaño del dataset de ajedrez:", chess_data.shape)
print("Tamaño del dataset de Bitcoin:", bitcoin_data.shape)
```

## Dataset de Partidas de Ajedrez
### Variables Potencialmente Interesantes:

**id**: Identificador único de cada partida. Esta variable es útil para distinguir entre diferentes partidas.

**crated_at**: Permite conocer la fecha y hora exacta en la que se creó la partida en el sistema.

**winner**: El resultado de la partida. Es una variable clave para entender los resultados.

**increment_code**: Tipo de control de tiempo utilizado en la partida. Esto puede afectar la calidad de las partidas y la toma de decisiones.

**white_id**: Nombre del jugador que juega con las piezas blancas. Este dato puede ser interesante para analizar las victorias por color de las piezas.

**white_rating**: Rating ELO del jugador de las blancas. Permite comparar el nivel de habilidad de los jugadores.

**black_id**: Nombre del jugador que juega con las piezas negras. Similar a player_white, se puede analizar el rendimiento por jugador o por color de piezas.

**black_rating**: Rating ELO del jugador de las negras. Similar a rating_white, esta variable es clave para comparar el rendimiento de los jugadores.

**moves**: Secuencia de movimientos realizados durante la partida, a menudo representados en notación algebraica. Esta variable puede ser útil para analizar patrones de juego y estrategias.

**opening_name**: Apertura utilizada en la partida. Las aperturas son un componente importante en el ajedrez y podrían ser analizadas para ver tendencias en la estrategia.

## Dataset de Cotización de Bitcoin
### Variables Potencialmente Interesantes:

**date**: La fecha de la cotización. Es fundamental para analizar la evolución del precio a lo largo del tiempo.

**open**: Precio de apertura de Bitcoin en un día determinado. Es útil para observar cómo se inicia el día con respecto a la tendencia previa.

**high**: Precio máximo alcanzado por Bitcoin durante el día. Esto puede dar una idea de la volatilidad y los picos de precios.

**low**: Precio mínimo alcanzado durante el día. Junto con high, se puede analizar la volatilidad y los rangos de precios.

**close**: Precio de cierre de Bitcoin al final del día. Es una de las variables más relevantes para evaluar el rendimiento de la criptomoneda a lo largo del tiempo.

**volume**: Volumen de Bitcoin negociado durante el día. Es importante para entender la liquidez y el interés en el mercado.


```python
# Revisar valores nulos en Ajedrez
print(chess_data.isnull().sum())

# Revisar valores nulos en Bitcoin
print(bitcoin_data.isnull().sum())
```

```python
import matplotlib.pyplot as plt
```

```python
# Analizar la relación entre el número de turnos y los ratings de los jugadores
plt.figure(figsize=(10,6))
plt.scatter(chess_data['turns'], chess_data['white_rating'], label='Rating del Blanco', color='r', alpha=0.6)
plt.scatter(chess_data['turns'], chess_data['black_rating'], label='Rating del Negro', color='g', alpha=0.6)
plt.xlabel('Número de Turnos')
plt.ylabel('Rating de los Jugadores')
plt.title('Relación entre Número de Turnos y Rating de los Jugadores')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Filtrar partidas donde el jugador de blancas y de negras tienen un ELO mayor o igual a 2000
chess_filtered = chess_data[(chess_data['white_rating'] >= 2000) & (chess_data['black_rating'] >= 2000)]
```

```python
# Obtener la cantidad de partidas después de los filtros
cantidad_partidas = chess_filtered.shape[0]

# Mostrar la cantidad de partidas
print(f"Cantidad de partidas filtradas: {cantidad_partidas}")
```

```python
# Mostrar las primeras filas del filtro
print(chess_filtered.head())
```

```python
# Filtrar además partidas de apertura peon rey
chess_filtered = chess_data[(chess_data['opening_eco'] == 'C20')]
```

```python
# Obtener la cantidad de partidas después de los filtros
cantidad_partidas = chess_filtered.shape[0]

# Mostrar la cantidad de partidas
print(f"Cantidad de partidas filtradas: {cantidad_partidas}")
```

```python
# Analizar la evolución del precio de Bitcoin:

# Convertir la columna 'date' a tipo datetime
bitcoin_data['date'] = pd.to_datetime(bitcoin_data['date'])

# Graficar el precio de cierre
plt.figure(figsize=(10,6))
plt.plot(bitcoin_data['date'], bitcoin_data['close'], label='Precio de Cierre de Bitcoin', color='c')
plt.xlabel('Fecha')
plt.ylabel('Precio en USD')
plt.title('Evolución del Precio de Bitcoin')
plt.legend()
plt.grid(True)
plt.show()
```
